## Example 1: Simple Beam Deflection

Calculate the deflection of a simply supported beam with a point load.

```python
# Define beam parameters symbolically
print("Beam parameters:")
print(f"Length: {L}")
print(f"Load position: {a} from left support")
print(f"Distance from right support: {b} = {L} - {a}")
print(f"Point load: {P}")
print(f"Young's modulus: {E}")
print(f"Moment of inertia: {I}")
```

```python
# Maximum deflection formula for a simply supported beam with point load
# When load is at midspan (a = L/2)
max_deflection = P * L**3 / (48 * E * I)
print("Maximum deflection (load at center):")
sp.pprint(max_deflection)

# General deflection at load point
deflection_at_load = P * a * b * (L**2 - b**2 - a**2) / (6 * E * I * L)
print("\nDeflection at load point:")
sp.pprint(deflection_at_load)
```

## Example 2: Units with Pint

Demonstrate the use of units for structural calculations.

```python
# Define concrete values with units
L_val = 10 * ureg.meter
P_val = 1000 * ureg.newton
E_val = 200e9 * ureg.pascal  # Steel modulus
I_val = 100e-6 * ureg.meter**4  # Moment of inertia

print("Beam properties with units:")
print(f"Length: {L_val}")
print(f"Load: {P_val}")
print(f"Young's modulus: {E_val}")
print(f"Moment of inertia: {I_val}")

# Calculate maximum deflection with units
max_def_val = P_val * L_val**3 / (48 * E_val * I_val)
print(f"\nMaximum deflection: {max_def_val.to('millimeter')}")
```

## Example 3: Plotting

Create a simple plot showing beam deflection shape.

```python
# Create a numerical example
L_num = 10  # meters
a_num = 4   # meters
P_num = 1000  # N
E_num = 200e9  # Pa
I_num = 100e-6  # m^4

# Create x values
x_vals = np.linspace(0, L_num, 100)

# Calculate deflection for each point (simplified for plotting)
# This is an approximation for visualization
deflections = []
for x_val in x_vals:
    if x_val <= a_num:
        # Left side of load
        v = P_num * (L_num - a_num) * x_val * (L_num**2 - (L_num - a_num)**2 - x_val**2) / (6 * E_num * I_num * L_num)
    else:
        # Right side of load
        v = P_num * a_num * (L_num - x_val) * (L_num**2 - a_num**2 - (L_num - x_val)**2) / (6 * E_num * I_num * L_num)
    deflections.append(v)

# Plot the beam deflection
plt.figure(figsize=(12, 6))
plt.plot(x_vals, np.array(deflections) * 1000, 'b-', linewidth=2, label='Beam deflection')
plt.axvline(x=a_num, color='r', linestyle='--', label=f'Load position (x={a_num}m)')
plt.xlabel('Position along beam (m)')
plt.ylabel('Deflection (mm)')
plt.title('Simply Supported Beam Deflection')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

print(f"Maximum deflection: {max(np.abs(deflections)) * 1000:.2f} mm")
```
